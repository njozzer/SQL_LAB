# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select tracks.name from invoice_items
join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
join tracks on (invoice_items.TrackId=tracks.TrackId)
join customers on (invoices.CustomerId=customers.CustomerId)
join employees on (employees.employeeid=customers.supportrepid)
where employees.hiredate > "2002-09-01"
```


## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select 
playlists.name as name_playlist,
sum(tracks.unitprice) as sum_money,
sum(tracks.milliseconds) as sum_time
from tracks
join playlist_track on(tracks.trackid=playlist_track.trackid)
join playlists on (playlists.playlistid=playlist_track.playlistid)
group by playlists.name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select artists.name,genres.name as genre_name ,count(tracks.name) as count from tracks
join genres on (tracks.genreid=genres.genreid)
join albums on (albums.albumid=tracks.albumid)
join artists on (albums.artistid=artists.artistid)
group by artists.artistid,genres.genreid
```

## Задание 4
Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql
with table1 as (
    select artists.name as artist_name,
    genres.name as genre_name,
    sum(tracks.unitprice) as sum_all_price
    from tracks
    join genres on (tracks.genreid=genres.genreid)
    join albums on (albums.albumid=tracks.albumid)
    join artists on (albums.artistid=artists.artistid)
    group by artists.name
),
table2 as(
select 
artists.name as artist_name,
genres.name as genre_name,
sum(tracks.unitprice) as sum_price
from tracks
join genres on (tracks.genreid=genres.genreid)
join albums on (albums.albumid=tracks.albumid)
join artists on (albums.artistid=artists.artistid)
group by artists.name,genres.name
),
table3 as(
    select artist_name,count(genre_name) as count_genre,max(sum_price) as max_sum_price from table2
    group by artist_name
)
select distinct table1.artist_name,
table1.genre_name,
(table1.sum_all_price-table3.max_sum_price)/table1.sum_all_price as result
from table1,table2,table3
where table1.artist_name=table2.artist_name
and table1.artist_name=table3.artist_name
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select customers.company as company,
count(tracks.trackid) as cnt 
from tracks
join genres on (tracks.genreid=genres.genreid)
join invoice_items on (invoice_items.trackid=tracks.trackid)
join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
join customers on (invoices.CustomerId=customers.CustomerId)
where genres.name = "Rock"
group by customers.company
```

## Задание 6
Вывести альбом, который является минимальным по объему(в байтах) среди альбомов со стоимостью, большей средней цены за альбом по всем альбомам с таким же жанром (если у альбома песни с нескольким жанрами, то жанром является тот, у которого наибольшее количество песен в альбоме)

```python
%%sql
with table1 as 
(select albums.title as album_title,genres.name as genre_name,count(tracks.trackid) as cnt from tracks
 join genres on (tracks.genreid=genres.genreid)
 join albums on (albums.albumid=tracks.albumid)
 group by albums.title,genres.name
)
select * from table1
```

```python
%%sql
WITH table1 AS (
    SELECT artists.ArtistId AS 'ArtistId', genres.GenreId AS 'GenreId', SUM(tracks.UnitPrice) AS 'Money'
    FROM artists
    INNER JOIN albums ON (artists.ArtistId = albums.ArtistId)
    INNER JOIN tracks ON (albums.AlbumId = tracks.AlbumId)
    INNER JOIN genres ON (tracks.GenreId = genres.GenreId)
    GROUP BY artists.ArtistId, genres.GenreId),
    table2 AS (
    SELECT table1.ArtistId AS 'ArtistId', MAX(table1.Money) AS 'Max', SUM(table1.Money) AS 'Sum'
    FROM table1
    GROUP BY table1.ArtistId),
    table3 AS (
    SELECT table1.ArtistId, table1.GenreId
    FROM table1, table2
    WHERE table1.ArtistId = table2.ArtistId AND
          table1.Money = table2.MAX),
    table4 AS (
    SELECT table1.ArtistId AS 'ArtistId', SUM(table1.Money) AS 'otherSum'
    FROM table1, table3
    WHERE table1.ArtistId = table3.ArtistId AND
         table1.GenreId != table3.GenreId
    GROUP BY table1.ArtistId)
SELECT artists.Name AS 'Artist', genres.Name AS 'Best genre', (table4.otherSum / table2.SUM) AS 'Koef'
FROM artists, genres, table1, table2, table3, table4
WHERE artists.ArtistId = table1.ArtistId AND
      artists.ArtistId = table2.ArtistId AND
      artists.ArtistId = table3.ArtistId AND
      artists.ArtistId = table4.ArtistId AND
      genres.GenreId = table3.GenreId
GROUP BY artists.ArtistId;
```
