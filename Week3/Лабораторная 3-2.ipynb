# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select tracks.name from invoice_items
join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
join tracks on (invoice_items.TrackId=tracks.TrackId)
join customers on (invoices.CustomerId=customers.CustomerId)
join employees on (employees.employeeid=customers.supportrepid)
where employees.hiredate > "2002-09-01"
```

Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select 
playlists.name as name_playlist,
sum(tracks.unitprice) as sum_money,
sum(tracks.milliseconds) as sum_time
from tracks
join playlist_track on(tracks.trackid=playlist_track.trackid)
join playlists on (playlists.playlistid=playlist_track.playlistid)
group by playlists.name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select artists.name,
genres.name as genre_name ,
count(tracks.name) as count 
from tracks
join genres on (tracks.genreid=genres.genreid)
join albums on (albums.albumid=tracks.albumid)
join artists on (albums.artistid=artists.artistid)
group by artists.name,genres.name
```

```python
%%sql
with table1 as(
    select artists.artistid as artist_id,
    artists.name as artist_name,
    genres.name as genre_name,
    count(tracks.name) as count 
    from tracks
    join genres on (tracks.genreid=genres.genreid)
    join albums on (albums.albumid=tracks.albumid)
    join artists on (albums.artistid=artists.artistid)
    group by artists.artistid,genres.genreid
)
select artist_name,genre_name,count from table1
```

Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql
with table1 as (
    select albums.artistid as artistID,genres.genreid as genreID, sum(tracks.unitprice) as money 
    from tracks
    join genres on (tracks.genreid = genres.genreid)
    join albums on (tracks.albumid = albums.albumid)
    
    group by albums.artistid,genres.genreid),
table2 as (
    select albums.artistid as artistID, sum(tracks.unitprice) as money_sum
    from tracks
    join albums on (tracks.albumid = albums.albumid)
    group by albums.artistid),
table3 as (
    select artistID,max(money) as max_money from table1
    group by artistID),
table4 as (
    select table1.artistID,table1.genreID, (table2.money_sum - table1.money)/table2.money_sum as res
    from table1
    join table3 on (table1.artistID=table3.artistID)
    join table2 on (table2.artistID=table1.artistID)
    where table3.max_money = table1.money)
select * from table4
```

```python
%%sql
with 
table0 as(
    select albums.artistid as artistID,
    genres.genreid as genreID,
    (invoice_items.unitprice *invoice_items.quantity) as money_per_track 
    from tracks
    join genres on (tracks.genreid = genres.genreid)
    join albums on (tracks.albumid = albums.albumid)
    join artists on (artists.artistid = albums.artistid)
    join invoice_items on (invoice_items.trackid=tracks.trackid)
),
table1 as (
    select artistID,
    genreID,
    sum(money_per_track) as money
    from table0
    group by artistid,genreid
),
table2 as (
    select artistID, 
    sum(money_per_track) as money_sum
    from table0
    group by artistid
),
table3 as (
    select artistID,
    max(money) as max_money 
    from table1
    group by artistID),
table4 as (
    select 
    table1.artistID as artistid,
    table1.genreID as genreid,
    (table2.money_sum - table1.money)/table2.money_sum as res
    from table1
    join table3 on (table1.artistID=table3.artistID)
    join table2 on (table2.artistID=table1.artistID)
    where table3.max_money = table1.money)


select artists.name as artist_name,
genres.name as genre_name,
table4.res from table4
join genres on (genres.genreid=table4.genreid)
join artists on (artists.artistid=table4.artistid)
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select customers.company as company,
count(tracks.trackid) as cnt 
from tracks
join genres on (tracks.genreid=genres.genreid)
join invoice_items on (invoice_items.trackid=tracks.trackid)
join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
join customers on (invoices.CustomerId=customers.CustomerId)
where genres.name = "Rock"
group by customers.company
```

## Задание 6
Вывести альбом, который является минимальным по объему(в байтах) среди альбомов со стоимостью, большей средней цены за альбом по всем альбомам с таким же жанром (если у альбома песни с нескольким жанрами, то жанром является тот, у которого наибольшее количество песен в альбоме)

```python
%%sql
with 
table1 as (
    select albums.albumid as album_id,
    sum(tracks.bytes) as album_size,
    sum(tracks.unitprice) as price
    from albums
    inner join tracks on (albums.albumid = tracks.albumid)
    group by albums.albumid
),
table2 as (
    select tracks.albumid as album_id,
    tracks.genreid as genre_id,
    count(tracks.trackid) as track_count
    from tracks
    group by tracks.albumid,tracks.genreid
),
table3 as (
    select table2.album_id as album_id,
    max(table2.track_count) as max_count
    from table2
    group by table2.album_id
),
table4 as(
    select table2.album_id as album_id,
    table2.genre_id as genre_id,
    table3.max_count as count
    from table2,table3
    where table3.album_id = table2.album_id and
    table2.track_count = table3.max_count
),
table5 as (
    select table4.genre_id as genre_id, 
    avg(table1.price) as avg_price_per_album
    from table1, table4
    where table1.album_id = table4.album_id
    group by table4.genre_id
),
table6 as(
    select table1.album_id,
    min(table1.album_size) as min_size,
    table5.avg_price_per_album 
    from table1,table4,table5
    where table1.album_id = table4.album_id and
    table4.genre_id = table5.genre_id and 
    table1.price > table5.avg_price_per_album
)
select 
albums.title,
table1.album_size,
table1.price,
table6.avg_price_per_album 
from table1,table6
join albums on (table1.album_id=albums.albumid)
where 
table1.album_id=table6.album_id and
table1.album_size=table6.min_size
```
