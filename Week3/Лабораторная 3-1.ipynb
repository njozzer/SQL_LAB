# Задание 1
Вывести 10 самых больших по размеру треков жанра ROCK и формата MPEG (исоп

```python
%%sql
select 
tracks.Name,
tracks.bytes
from tracks
inner join genres on (genres.GenreId=tracks.GenreId)
inner join media_types on (media_types.MediaTypeId=tracks.MediaTypeid)
where genres.name = "Rock" and  media_types.name = "MPEG audio file"
order by tracks.bytes desc limit 10
```

## Задание 2
Вывести названия всех групп, их песен и названия их альбомов для всех треков жанра Рок, приобретенные сотрудниками Microsoft.

```python
%%sql
select 
artists.name as artist_name,
tracks.name as track_name,
albums.title as album_title
from invoice_items
inner join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
inner join tracks on (invoice_items.TrackId=tracks.TrackId)
inner join customers on (invoices.CustomerId=customers.CustomerId)
inner join albums on (albums.AlbumId=tracks.AlbumId)
inner join artists on (albums.ArtistId=artists.ArtistId)
inner join genres on (genres.GenreId=tracks.GenreId)
where genres.name="Rock" and customers.Company="Microsoft Corporation"
```

## Задание 3
Для каждого набора (жанр, тип медиа) вывести среднюю цену по стоимости трека и общее количество, причем вывести только те наборы, для которых все треку стоят больше 1,5$.


```python
%%sql
select 
media_types.name as media_type,
genres.name as genre_name,
avg(tracks.unitprice) as avg_price,
count(*) cnt
from tracks
inner join genres on (genres.GenreId=tracks.GenreId)
inner join media_types on (media_types.MediaTypeId=tracks.MediaTypeId)
group by media_types.name,genres.name
having min(tracks.unitprice)>1.5
```

## Задание 4
Вывести компании, сделавшие максимальное и минимальное число заказов.

```python
%%sql
with table1 as (
select
Customers.company as company_name,count(invoices.invoiceid) as cnt
from invoices
inner join customers on (invoices.CustomerId=customers.CustomerId)
group by company_name),
table2 as (
select min(cnt) as min,max(cnt) as max from table1
)
select company_name,cnt from table1,table2
where cnt = min or cnt = max
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру поп-музыки

```python
%%sql
select 
Customers.company as company_name,
count(*) cnt
from invoice_items
inner join invoices on (invoices.InvoiceId=invoice_items.InvoiceId)
inner join tracks on (invoice_items.TrackId=tracks.TrackId)
inner join genres on (tracks.Genreid=genres.Genreid)
inner join customers on (invoices.CustomerId=customers.CustomerId)
where Genres.name="Pop"
group by Customers.company
```

## Задание 6
Вывести средний размер альбома в байтах.

```python
%%sql
select avg(sum_tr) as avg_bytes
from
(select sum(tracks.Bytes) as sum_tr
from albums
join tracks on (albums.AlbumId = tracks.AlbumId)
group by albums.AlbumId) ;
```
