# Описание таблиц

## employees
Таблица employees хранит данные сотрудников, такие как employee id, last name, fist name, и т.д. Также есть поле _ReportsTo_, которое показывает, кто чей подчиненный

```python
%%sql 
select * 
from employees
```

## Таблица customers
Таблица customers хранит данные по клиентам. Поле SupportRepid является внешним ключом к таблице employees

```python
%%sql
SELECT * 
FROM CUSTOMERS
```

## Таблицы invoices и invoice_items 
Эти две таблицы хранят информацию по заказам. Таблица Invoice хранит заголовок заказа, а также внешний ключ к таблице customers. Таблица invoice_items хранит детали заказа, а также внешний ключ к таблицам invoices, tracks

```python
%%sql
SELECT * 
FROM invoices
```

```python
%%sql
SELECT * 
FROM invoice_items
```

## Таблица artists
Таблица артистов хранит данные об артистах (в данном случае только id и name)

```python
%%sql
SELECT * 
FROM artists
```

## Таблица albums
Таблица albums хранит информацию об альбомах, в ней есть 2 внешних ключа - на таблицу artists и tracks

```python
%%sql
SELECT * 
FROM albums
```

## Таблица media_types
Таблица media_types хранит информацию о возможных форматах

```python
%%sql
SELECT * 
FROM media_types
```

## Таблица genres 
Таблица genres хранит информацию о всевозможных форматах таблиц

```python
%%sql 
select * 
from genres
```

## Таблица tracks 
Таблица tracks хранит информации о композициях. В данной таблице есть внешние ключи на albums, mediatypes и genres

```python
%%sql
SELECT * 
FROM tracks
```

## Таблицы playlists и playlist_track
Таблица playlists хранит список плейлистов. Таблица playlist_track - таблица связи многие-ко-многим

```python
%%sql
SELECT * 
FROM playlists
```
